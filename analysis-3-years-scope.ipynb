```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#importing SEPA/EEA dataset
dataset = pd.read_csv('DataExtract.csv')
```

```python
df_dataset = pd.DataFrame(data=dataset)
df_dataset.head()
```

```python
#deleting columns with redundant data
df_drop = df_dataset.drop(columns=['Country', 'Sampling Point Id', 'Data Aggregation Process Id', 'Data Aggregation Process', 'City', 'City Code', 'City Population', 'Link to raw data (only E1a/validated data from AQ e-Reporting)'])


#keeping only data from 2019, 2020 and 2021
df_drop.drop(df_drop[df_drop['Year'] < 2019].index, inplace = True)
df_drop.drop(df_drop[df_drop['Year'] > 2021].index, inplace = True)
```

```python
dfbase = df_drop
```

```python
#removing 6 stations that are present only in 2021. year  
dffix = df_drop.drop(df_drop.index[dfbase['Air Quality Station Name'].isin(['Beocin-Centar', 'Kosjeric', 'Novi Pazar', 'Obrenovac-Centar', 'Popovac', 'Uzice'])])


#test 
locations_check = dffix['Air Quality Station Name'].unique()
print(sorted(locations_check))
```

```python
#saving df as .csv file
dffix.to_csv("19-21.csv", sep='\t', encoding='utf-8')
```

```python
#extracting pollutants 
CO = dffix.loc[dffix['Air Pollutant'] == "CO"]
NO2 = dffix.loc[dffix['Air Pollutant'] == "NO2"]
O3 = dffix.loc[dffix['Air Pollutant'] == "O3"]
PM10 = dffix.loc[dffix['Air Pollutant'] == "PM10"]
SO2 = dffix.loc[dffix['Air Pollutant'] == "SO2"]
```

```python
#saving dataframes (pollutants) as separate .csv files for further usage
CO.to_csv("CO-19-21.csv", sep='\t', encoding='utf-8')
NO2.to_csv("NO2-19-21.csv", sep='\t', encoding='utf-8')
O3.to_csv("O3-19-21.csv", sep='\t', encoding='utf-8')
PM10.to_csv("PM10-19-21.csv", sep='\t', encoding='utf-8')
SO2.to_csv("SO2-19-21.csv", sep='\t', encoding='utf-8')
```

```python
#CO means and max values (per locations)
CO_average = CO.groupby('Air Quality Station Name')['Air Pollution Level'].mean()
top_CO_average = CO_average.nlargest(10)

CO_max = CO.groupby('Air Quality Station Name')['Air Pollution Level'].max()
top_CO_max = CO_max.nlargest(10)

co_average_values_sorted = pd.DataFrame(top_CO_average)
co_max_values_sorted = pd.DataFrame(top_CO_max)
```

```python
#izbacivanje 6 lokacija zabelezenih unutar 2021
dfbase1 = dfbase.drop(dfbase["Air Quality Station Name"].loc[dfbase["Air Quality Station Name"]=='Beocin-Centar'].index)
dfbase2 = dfbase1.drop(dfbase1["Air Quality Station Name"].loc[dfbase1["Air Quality Station Name"]=='Kosjeric'].index)
dfbase3 = dfbase2.drop(dfbase2["Air Quality Station Name"].loc[dfbase2["Air Quality Station Name"]=='Novi Pazar'].index)
dfbase4 = dfbase3.drop(dfbase3["Air Quality Station Name"].loc[dfbase3["Air Quality Station Name"]=='Obrenovac-Centar'].index)
dfbase5 = dfbase4.drop(dfbase4["Air Quality Station Name"].loc[dfbase4["Air Quality Station Name"]=='Popovac'].index)
dfbase6 = dfbase5.drop(dfbase5["Air Quality Station Name"].loc[dfbase5["Air Quality Station Name"]=='Uzice'].index)

dffix = dfbase6
dffix
```

```python
#extracting pollutants SUBSETS
CO = dffix.loc[dffix['Air Pollutant'] == "CO"]
NO2 = dffix.loc[dffix['Air Pollutant'] == "NO2"]
O3 = dffix.loc[dffix['Air Pollutant'] == "O3"]
PM10 = dffix.loc[dffix['Air Pollutant'] == "PM10"]
SO2 = dffix.loc[dffix['Air Pollutant'] == "SO2"]
```

```python
#df po godinama
CO.to_csv("CO-19-21.csv", sep='\t', encoding='utf-8')
NO2.to_csv("NO2-19-21.csv", sep='\t', encoding='utf-8')
O3.to_csv("O3-19-21.csv", sep='\t', encoding='utf-8')
PM10.to_csv("PM10-19-21.csv", sep='\t', encoding='utf-8')
SO2.to_csv("SO2-19-21.csv", sep='\t', encoding='utf-8')
```

```python
CO_means = CO.groupby('Air Quality Station Name')['Air Pollution Level'].mean()
top_CO_means = CO_means.nlargest(10)

CO_max = CO.groupby('Air Quality Station Name')['Air Pollution Level'].max()
top_CO_max = CO_max.nlargest(10)


cotest = pd.DataFrame(top_CO_means)
cotest2 = pd.DataFrame(top_CO_max)
```
